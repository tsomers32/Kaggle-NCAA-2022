```python
import numpy as np
import pandas as pd
import keras
import tensorflow
```

```python
train_data = pd.read_csv('C:/Users/Travis/Desktop/Kaggle-NCAA-2022/training_data.csv', dtype = np.float64)
test_data = pd.read_csv('C:/Users/Travis/Desktop/Kaggle-NCAA-2022/test_data.csv', dtype = np.float64)
```

```python
train_data.head()
```

```python
x_train = train_data.iloc[:,2:]
y_train = train_data.team1Win
x_test = test_data.iloc[:,2:]
y_test = test_data.team1Win
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Dense(35, input_dim = 98, activation = 'relu'))
model.add(keras.layers.Dropout(.5))
model.add(keras.layers.Dense(15, activation = 'relu'))
model.add(keras.layers.Dropout(.5))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy', 'binary_crossentropy'])
```

```python
x_tensor = tensorflow.convert_to_tensor(x_train)
y_tensor = tensorflow.convert_to_tensor(y_train)
```

```python
model.fit(x_tensor, y_tensor, epochs = 150, batch_size = 5)
```

```python
model.evaluate(x_test, y_test)
```

```python
pd.DataFrame(model.predict(x_train)).to_csv('C:/Users/Travis/Desktop/Kaggle-NCAA-2022/nn_train_pred.csv',index=False)
```

```python
pd.DataFrame(model.predict(x_test)).to_csv('C:/Users/Travis/Desktop/Kaggle-NCAA-2022/nn_test_pred.csv',index=False)
```
